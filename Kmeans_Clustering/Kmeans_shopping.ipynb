```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
spending_index = [3, 3, 5, 2, 4, 6, 6, 6, 5, 6, 7, 8, 9, 8, 9, 9, 8]
income_index = [5, 4, 6, 6, 5, 8, 2, 3, 6, 7, 2, 5, 1, 2, 6, 1, 3]

data = list(zip(spending_index, income_index))
```

```python
def generate_clusters(n_clusters: str):
    model = KMeans(n_clusters=n_clusters)
    clusters = model.fit_predict(data)
    sample_colors = ['red', 'green', 'blue', 'yellow']
    colors = [sample_colors[label] for label in clusters]
    plt.scatter(x=spending_index, y=income_index, c=colors)
    plt.title(f'KMeans clustering - {n_clusters} clusters')
    fig = plt.gcf()
    fig.set_size_inches(5, 3)
```

```python
generate_clusters(n_clusters=2)
```

```python
generate_clusters(n_clusters=3)
```

```python
generate_clusters(n_clusters=4)
```
