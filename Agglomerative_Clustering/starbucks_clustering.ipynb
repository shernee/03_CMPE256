```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
```

```python
pt = ['6am-7am', '7am-8am', '8am-9am', '9am-10am', '10am-12pm', '12pm-1pm', '1pm-3pm', '3pm-4pm', '4pm-7pm']
hb = ['Coffee Type', 'Cafe Latte', 'Cafe Mocha', 'Cappuccino', 
'Caramel Macchiato', 'Cafe Americano', 'Hot Chocolate', 'Espresso Shot', 'Coffee of the Day']

xticks = [i for i in range(1, len(pt)+1)]
yticks = [i for i in range(1, len(hb)+1)]

pt_x = [1, 1, 2, 2, 3, 3, 3, 3, 4, 4, 3, 5, 7, 5, 4, 6, 7, 7, 8, 8]
hb_y = [5, 7, 6, 9, 6, 8, 8, 9, 8, 8, 3, 4, 2, 7, 5, 9, 3, 9, 1, 7]

data = list(zip(pt_x, hb_y))
```

```python
def generate_clusters(linkage: str, n_clusters: str = 2):
    model = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage)
    clusters = model.fit_predict(data)

    sample_colors = ['red', 'green', 'blue', 'yellow', 'purple', 'orange']
    colors = [sample_colors[label] for label in clusters]
    plt.scatter(x=pt_x, y=hb_y, c=colors)
    plt.xlabel = 'Popular times'
    plt.ylabel = 'Hot Beverage'
    plt.xticks(ticks=xticks, labels=pt, rotation=45)
    plt.yticks(ticks=yticks, labels=hb)
    plt.title(f'{linkage} link clustering')
    fig = plt.gcf()
    fig.set_size_inches(5, 3)
    
```

```python
# linkage = 'single', 'average', 'complete'

#generate_clusters(linkage='single')
generate_clusters(linkage='single', n_clusters=3)
```

```python
generate_clusters(linkage='average', n_clusters=3)
```

```python
generate_clusters(linkage='complete', n_clusters=3)
```
