```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
x = [2, 2, 3, 5, 5, 6, 5, 7, 8, 10, 12]
y = [2, 6, 7, 8, 5, 6, 2, 3, 4, 6, 8]

data = list(zip(x,y))
```

```python
def generate_clusters(linkage: str, n_clusters: str = 2):
    model = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage)
    clusters = model.fit_predict(data)

    sample_colors = ['red', 'green', 'blue', 'yellow', 'purple', 'orange']
    colors = [sample_colors[label] for label in clusters]
    plt.scatter(x=x, y=y, c=colors)
```

```python
# linkage = 'single', 'average', 'complete'

#generate_clusters(linkage='single')
generate_clusters(linkage='single', n_clusters=5)
```

```python
generate_clusters(linkage='average', n_clusters=3)
```

```python
generate_clusters(linkage='complete', n_clusters=4)
```
